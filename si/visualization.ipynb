```python
# import numpy as np
# from tqdm.notebook import tqdm
from matplotlib import pyplot as plt
import seaborn as sns
# import pandas as pd
# import numpy_indexed as npi
# from scipy.io import loadmat


# from src.ipm import NeuronReader, grablabels, NeuronLoader
# from src.metrics import sepind
# from src.utils import *

# from imblearn.under_sampling import RandomUnderSampler as rus
# from scipy.ndimage import gaussian_filter1d as smooth

# from sklearn.covariance import EmpiricalCovariance
# from sklearn.decomposition import PCA

import os
import sys
import json

with open('../dirs.json', 'r') as f:
    dirs = json.load(f)
sys.path.append(dirs['root'])

import numpy as np
```

```python
outPath = dirs['out']['sep']

midItc = np.load(os.path.join(outPath, 'both-fast-it-mid.npy'))
supItc = np.load(os.path.join(outPath, 'both-fast-it-sup.npy'))
midPfc = np.load(os.path.join(outPath, 'both-fast-pfc-mid.npy'))
supPfc = np.load(os.path.join(outPath, 'both-fast-pfc-sup.npy'))

time = np.arange(-200, 650) + 25
```

```python
fig, axs = plt.subplots(2, 2, figsize=(16, 6))
axs = axs.flatten()
axs[0].plot(time, midItc.T, lw=1, c='gray', alpha=.35);
axs[0].plot(time, midItc.mean(0), lw=5);
axs[0].set_title("Mid, ITC")

axs[1].plot(time, supItc.T, lw=1, c='gray', alpha=.35);
axs[1].plot(time, supItc.mean(0), lw=5);
axs[1].set_title("Sup, ITC")

axs[2].plot(time, midPfc.T, lw=1, c='gray', alpha=.35);
axs[2].plot(time, midPfc.mean(0), lw=5);
axs[2].set_title("Mid, PFC")

axs[3].plot(time, supPfc.T, lw=1, c='gray', alpha=.35);
axs[3].plot(time, supPfc.mean(0), lw=5);
axs[3].set_title("Sup, PFC")

for ax in axs:
    ax.set_xlim(-100, 650)

sns.despine()
fig.tight_layout()
```

```python
blIndex = (time > -50) & (time < 0)

fig, axs = plt.subplots(1, 2, figsize=(16, 4), sharey=True)
axs = axs.flatten()

axs[0].plot(time, midItc.mean(0) - midItc[:, time < blIndex].mean(), lw=5);
axs[0].plot(time, supItc.mean(0) - supItc[:, time < blIndex].mean(), lw=5);
axs[0].set_title("ITC", fontsize=20)

axs[1].plot(time, midPfc.mean(0) - midPfc[:, time < blIndex].mean(), lw=5);
axs[1].plot(time, supPfc.mean(0) - supPfc[:, time < blIndex].mean(), lw=5);
axs[1].set_title("PFC", fontsize=20)

for ax in axs:
    ax.set_xlim(-100, 650)
    ax.legend(["Mid", "Sup"], frameon=False, fontsize=14)

sns.despine()
fig.tight_layout()
```
